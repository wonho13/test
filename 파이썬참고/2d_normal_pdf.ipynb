```python
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

def normal_pdf_2d(theta,mean,cov):
    return multivariate_normal.pdf(theta,mean=mean,cov=cov)

x, y = np.mgrid[-3:3:.01, -3:3:.01]
theta = np.empty(x.shape + (2,))
theta[:, :, 0] = x; theta[:, :, 1] = y;
z = normal_pdf_2d(theta,mean=np.zeros(2),cov=np.identity(2));
plt.contourf(x, y, z);
```

```python
fig = plt.figure(figsize=(10,8));
ax = fig.add_subplot(111,projection='3d');
ax.view_init(25,45);
ax.set_xlim(3,-3);
ax.plot_surface(x,y,z,cmap='hsv',edgecolor='gray');
```
