## 5 Sequential Monte Carlo methods

2장에서는 일반적인 상태 공간 모델(proposition 2.1)에 대한 재귀 필터링에 대해 소개하였습니다. 시간 $t-1$에서의 필터링 분포와 관측치 $y_t$로부터 시간 $t$에서의 필터링 분포를 계산하는 이러한 알고리즘의 재귀적 특성은 실시간 추론이 필요한 분야에 적합합니다. 이러한 분야에서는 현재 시스템 상태에 대한 최신 정보를 추론해야 합니다. 이러한 실시간 추론의 적용 사례들은 다음과 같습니다: 레이더로 관측한 항공기의 위치 및 속도 추정; 인공위성 데이터 기반의 태풍의 위치 및 특성 감시; 실시간 데이터 기반의 주식 가격 변동 추정. 불행히도, 일반적인 상태공간 모델의 경우 (2.7)의 적분을 정확히 구하는 것은 불가능합니다. DLM은 칼만 필터를 사용해 필터링 문제에 대한 닫힌 해를 계산할 수 있는 특수한 경우 입니다. 하지만 이 경우에도, DLM이 알지 못하는 변수를 포함하고 있으면, 일부 특수한 경우(4.3.1장 참고)를 제외하면 칼만 필터만으로는 필터링 분포를 계산할 수 없습니다.

off-line 또는 batch 조건인 경우, 4장에서 설명한 것과 같이 MCMC 방법을 사용하면 알지 못하는 변수를 포함한 DLM에 대한 추론이 가능하며 비선형, 비가우시안 모델에도 적용 가능합니다. 하지만, 이러한 방법은 실시간(online) 추론에 사용하기 어렵습니다. 왜냐하면 새로운 관측치를 얻게 되면 완전히 새로운 마르코프 체인을 시뮬레이션 해야 하기 때문입니다. 다시말해, MCMC 방법에서 $t-1$의 관측치를 사용해 추론한 결과는 $t$의 관측치를 사용한 사후 분포를 평가하는데 사용할 수 없습니다. 이러한 관점에서, 칼만필터와 달리, MCMC는 순차적인 계산에 적용하기 쉽지 않습니다.

비선형 및 비가우시안 상태공간 모델에 대한 필터링 분포를 순차적으로 업데이트 하려는 초기의 시도들은 상태공간 방정식 및 시슽메 방정식을 선형화 한 형태와 가우시안 추정을 기반으로 하였습니다 (상세 내용은 Cappe 외;2007을 참고). 이러한 방법은 비선형성이 약한 모델에는 적절했지만 비선형성이 강한 모델에는 좋지 않은 성능을 보였습니다. 또한 단순한 DLM의 경우에도 알지 못하는 변수를 순차적으로 추론하는 문제를 해결하지 못했습니다.

본 장에서는, 순차 몬테 카를로라고 불리는 비교적 최신 시뮬레이션 방법을 소개할 것입니다. 이 방법은 알지못하는 변수를 포함하는 DLM과 비선형 비가우시안 상태공간 모델에 적용했을 경우 매우 좋은 성능을 보이는 것이 입증되었습니다. 순차 몬테 카를로 방법은 복잡한 사후 분포에 대한 대뮬레이션 기반 알고리즘의 대안이 될 수 있습니다. 이 방법은 시간이력 모델에만 국한되지는 않지만, DLM과 사후 분포에 대한 업데이트가 자주 필요한 상태공간 모델의 경우 매우 우수한 성능을 보이는 것이 입증되었습니다. 순차 몬테 카를로 방법에 대한 연구는 현재 매우 활발히 이루어지고 있으며, 이 분야에 대한 자세한 내용들은 다루지 않을 것입니다. 대신, 일반적인 소개와 DLM에서 구현하기 쉬운 일부 알고리즘들에 대한 좀더 구체적인 내용을 소개할 것입니다. 이 분야에 관심이 있으신 분들은 Liu(2007), Doucet 외(2001), Del Moral(2004), Cappe 외(2005)의 책들을 참고하시기 바랍니다. Cappe 외(2007)의 논문은 이 분야에 대한 전반적인 현황에 대해 소개합니다.
