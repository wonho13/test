```python
import warnings
warnings.filterwarnings(action='ignore');
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
```

```python
# 이 논문 식으로 해보니 결과기 이상해 Dirlik 논문 식 사용함.
def w(z,r):
    c = np.sqrt(1-r**2);
    return (c/(np.sqrt(2*np.pi)))*np.exp(-(z**2)/(2*c**2))+ r*z*np.exp(-(z**2)/2)*(1/2+(erf((r*z/c)/np.sqrt(2))/2));
# 논문의 erf와 python erf의 수식이 달라 wiki를 참고해 수정하였음.
```

```python
z = np.linspace(-3,4,1000);
```

```python
ww = []; r = np.array([0,.2,.4,.6,.8,1]);
lines = ["-","--","-.",":","-","."];
plt.figure(figsize=(8,6));
for i in range(len(r)):
    ww.append(w(z,r[i]));
    plt.plot(z,ww[i],lines[i],label="$\gamma$=%.1f"%r[i]);
    plt.ylim(0,0.7);
plt.legend(loc='upper right');
plt.xlabel('$Z=\frac{S_a}{\sigma(S_a)}$');
plt.ylabel('$f(S_a)$');
```
