```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#input, output : abs acc
def H(f,fn,zeta):
    r = f/fn;
    A = 1 + (2*zeta*r)**2
    B = (1-r**2)**2 + (2*zeta*r)**2
    return np.sqrt(A/B);
    
```

```python
f   = np.linspace(1,1000,1000);
fn  = 100;
zeta = np.linspace(0.01,0.05,5);
```

```python
plt.figure(figsize=(20,10));
for z in zeta:
    plt.plot(f,H(f,fn,z),label='%.2f'%(z))
plt.xscale('log');
plt.yscale('log');
plt.legend();
plt.grid(which='both');
```
