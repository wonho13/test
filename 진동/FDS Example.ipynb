**Example 9.3**

An electronic system mounted directly on a vehicle body(a sprung mass) is subjected to road-load driving, inducing random vibration to the system. It is recommended that the vehicle manufacturer and supplier perform the random vibration test based on the ISO (International Organization for Standardization) 16750-3 standard.

The test duration shoud be 8 hours for each principal axis of the system. The power spectral density function of the base random excitation to the system is shown in Figure 9.13 and tabulated in Table 9.3.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.array([10,55,180,300,360,1000]);
p = np.array([0.208,0.0677,0.0026,0.0026,0.00146,0.00146]);
plt.plot(f,p,'k-'); plt.xscale('log'); plt.yscale('log');
plt.xlabel('Frequency (Hz)'); plt.ylabel('Input ($G^2/Hz$)');
plt.xlim(10,1000);
plt.grid(which='both'); plt.title('Figure 9.13 PSD plot based on the ISO 16750-3 random vibration profile');
```

It is assumed that the S-N curve follows $N_f\cdot S_a^m=C$ where $m=4$ and $C=1$ and that the linear relationship between the stress amplitude and the relative displacement exists as $Sa=K\cdot z$ where $K=10^6$. The $Q=10$ for dynamic amplification factor is assumed, equivalent to the damping ratio of $\zeta=5$%.

Calculate the fatigue damage of the electronic system based on Wirshing-Light method, the Ortiz-Chen method, and the Dirlik method.

**상수정리**

K : 변위에 의한 응력은 K로 정비례관계임을 가정

Q : 증폭비 ($1/2Q=\zeta$)

m, C : S-N 커브와 관련된 계수

**Solution**

1. Narrow-Band Damage

$$D_{NB}=\frac{E[0^+]T}{C}E[S_a^m]=\frac{E[0^+]T}{C}(\sqrt{2M_0})^m \Gamma(\frac{m}{2}+1)$$

where,

$E[0^+]=\sqrt{\frac{M_2}{M_0}}$

$M_j=\int_0^\infty f^j W_{S_a}(f) df$

$W_{S_a}(f)$ : stress RPSD
