```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
from scipy.signal import welch
import warnings
warnings.filterwarnings(action='ignore');
```

```python
# Rectangular PSD
def Gr(f,f1,p1,f2,p2,A1,A2):
    out = np.zeros(len(f));
    for i in range(len(f)):
        if f[i] >= f1 and f[i] <= p1:
            out[i] = A1
        elif f[i] >= f2 and f[i] <= p2:
            out[i] = A2;
    return out;
```

```python
# Smooth PSD
def Gs(f,A1,f1,Q1,A2,f2,Q2,fc):
    out1 = np.zeros(len(f)); out2 = np.zeros(len(f));
    for i in range(len(f)):
        if f[i] <= fc:
            out1[i] = A1/np.sqrt(1+((f1-f[i])**2)/(Q1**2));
            out2[i] = A2/np.sqrt(1+((f2-f[i])**2)/(Q2**2));
        else:
            out1[i] = 0;
            out2[i] = 0;
    out = out1 + out2;
    return out
```

```python
def psd2th(freq,psd,dur):
    fmax = max(freq); fs = 2*fmax; dt = 1/fs; tmax = dur;
    Nt = 2**int(np.log(int(dur/dt))/np.log(2));
    tmax = dt*Nt; df = 1/tmax; Nf = int(Nt/2); fmax=df*Nf;
    
    time = np.arange(Nt)*dt;
    
    fi = np.arange(Nf)*df;
    pi = 10**np.interp(np.log10(fi),np.log10(freq),np.log10(psd));

    g = np.zeros(Nt,'complex');
    # IFFT Method
    phi = np.random.uniform(low=-np.pi,high=np.pi,size=Nf);
    g[:Nf] = np.sqrt(pi)*np.exp(1j*phi);
    g[Nf:] = np.flip(g[:Nf]);
    g[Nf-1] = 0;
    g[Nf] = 0;
    th = np.fft.fft(g);
    th = th.real

    grms_spec = np.sqrt(np.trapz(pi,fi));
    grms_syn = np.std(th)

    th *= grms_spec/grms_syn;
    
    return time,th
```

```python
def psdinteg(freq,psd):
    f = freq; y = psd;
    ndata = len(y);
    a = np.zeros(ndata-1);
    for i in range(ndata-1):
        n = np.log10(y[i+1]/y[i])/np.log10(f[i+1]/f[i]);
        if n == -1:
            a[i] = y[i]*f[i]*np.log(f[i+1]/f[i]);
        else:
            a[i] = (y[i]/(f[i]**n))*(1/(n+1))*(f[i+1]**(n+1)-f[i]**(n+1));
    return np.sqrt(sum(a))
```
