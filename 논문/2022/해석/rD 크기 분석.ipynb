1 Hz, 200 Hz, 500 Hz, 1000 Hz에서 어떤 일이..?

Band1 또는 Band2의 넓이가 Full Band와 비슷할수록 rD는 1에 가까워 진다.

즉, A1/A2 가 1에 가까울수록 rD는 2로 근접해간다.

```python
fi = np.linspace(1,1000,5000); pi = np.zeros(5000);
for i in range(len(f)):
    pi[i] = blwn(fi[i],fmin,fmax,bwp,loc);
dfi,dpi = band_split(fi,pi,2);
K = 1
zeta = z;
```

```python
fn = np.array([1,200,500,1000]);
for i in range(len(fn)):
    rpsd = pi*HH(fi,fn[i],zeta)*K;
    rpsd1 = dpi[0]*HH(dfi[0],fn[i],zeta)*K;
    rpsd2 = dpi[1]*HH(dfi[1],fn[i],zeta)*K;
    # M0, M2, M4 넓이: 모두 한쪽으로 치우치면 rD가 작아진다.
    plt.figure(figsize=(12,3));
    plt.subplot(131); plt.plot(dfi[0],rpsd1); plt.plot(dfi[1],rpsd2);
    plt.subplot(132); plt.plot(dfi[0],rpsd1*dfi[0]**2); plt.plot(dfi[1],rpsd2*dfi[1]**2);
    plt.subplot(133); plt.plot(dfi[0],rpsd1*dfi[0]**4); plt.plot(dfi[1],rpsd2*dfi[1]**4);
```
