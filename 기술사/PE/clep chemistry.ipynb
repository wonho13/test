## Atomic theory and atomic structure

### Evidence for the atomic theory

(1) JJ Thomson 실험

유리관에 금속 전극을 떨어트려 놓고 배터리 연결. 음극선이 특정 방향으로 휘어짐. 입자는 전하를 띄고, 질량이 있다고 결론.

(2) Millikan 실험

oil drop 실험으로 중력과 전기력이 같도록 하여 질량과 전하량 구함. 

(3) Rutherford 실험

금박에 알파입자를 쏴 대부분은 통과하고 일부는 반사됨을 확인함. 아주 작은 핵(양성자)갖 중심에 있고 그 질량은 매우 큼을 확인.

### Atomic masses; determination by chemical and physical means

동위원소(isotope)는 양성자 수(원자번호)는 같지만 중성자 수는 다름.

$$ ^{M=Z+N}_{Z} A$$

질량수는 양성자수(Z)+중성자수(N) 임.

1amu는 원자 질량 단위임. $^{12}_6 C$는 양성자수6개, 중성자수6개, 질량은 12 amu 임.

원자량(atomic weight, 주기율표에 있는 값)는 자연계 존재하는 원자 질량을 가중평균한 것임. 예를들어 12가 99%, 13이 1% 있으면 원자량은 12(0.99)+13(0.01)임.

### Atomic number and mass number; isotopes and mass spectroscopy

mass spectroscopy : 원자를 쏘면 질량에 따라 꺾이는 정도가 다르기 때문에 이를 통해 질량 분포를 확인하는 장치


### Electron energy levels: atomic spectra, atomic orbitals

EM Spectrum은 다양한 주파수 및 파장을 가짐

$E=h\nu=hc/\lambda$

h : 플랑크 상수

$\nu$ : 주파수 (에너지는 주파수에 비례)

c : 빛의 주파수

$\lambda$ : 빛의 파장

Spectroscopy는 빛과의 상호 관계를 보여줌

[양자번호]

(1) 주양자번호(principal quantum number, n)

n = 1,2,3...

(2) 각모멘텀양자번호(angular momentum quatum number, l)

l < n-1

l=0 (s orbital), l=1 (p orbital), l=2 (d orbital), l=3 (f orbital)

s : sphere, 나머지는 pdf로 암기

(3) 자기양자번호(magnetic quantum nubmer, $m_l$)

l=0 ($m_l=0$), l=1 ($m_l=-1,0,+1$), l=2 ($m_l=-2,-1,0,+1,+2$)

(4) 회전양자번호(spin quantum number, $m_s$)

전자의 회전 방향 구분

$m_s= +1/2 \;or -1/2$


* 각 orbital(s,p,d,f...)에 포함될 수 있는 최대 전자 개수는 2개임

* 각 쉘(n=1,2,3...)에 포함될 수 있는 orbital 수는 2, 6, 6, 6...임 -> 1s2 2s2 2p6 3s2 3p6 3f6 ...


### Periodic relationships, including, for example, atomic radii, ionization energies, electron affinities, and oxidation states

[주기율표 구분]

(1) Main Group

주기율표 왼쪽, 오른쪽

Valence는 s와(왼쪽원자들) p만(He 제외한 오른쪽 원자들) 있음

s block(왼쪽 원자들 + He) : alkali metals, alkali earth metals

p block(He 제외한 오른쪽 원자들) : metals, metalloids, nonmetals (including halogens and noble gases)	

(2) Transition Metals

주기율표 중간 및 아래

d block(주기율표 중간) : transition metal

f block(주기율표 아래(별도)) : rare earth (lanthanide/actinide)


* 주기율표에서 멀고 다른 그룹이면 이온결합, 가깝고 같은 그룹이면 공유결합

[Periodic Relationship]

주기성은 최외곽(valence) 전자, 내부(core) 전자, 양자(proton) 수로 설명 가능

(1) Atomic radii

주기가 증가함에 따라(오른쪽으로 갈수록) 감소하며, 그룹에 따라(아래로 갈수록) 증가함 (Main Group만 해당)

* Ionic Radius : 이온 간의 거리. 금속의 경우 결합 시 최외곽 전자를 잃어 원자반경보다 작고, 비금속의 최외곽 전자를 얻으므로 경우 원자반경보다 큼

(2) Ionization energy

기체 상에서, ground state로부터 전자를 제거하기 위해 필요한 에너지

$A(g) \rightarrow A^{+}(g)+e^{-}$

(3) Electron Affinity

기체 상에서 전자를 추가해 ground state로 가기 위해 필요한 에너지

$A(g) + e^{-} \rightarrow A^{-}(g)$

### Kinetic-molecular theory

기체는 완전 구인 입자로 구성됨

기체입자는 일정한 무작위 거동을 함

기체입자는 직선으로 움직임

기체입자는 매우 멀리 떨어져 있음

**온도**는 평균운동에너지에 비례함

기체입자는 **독립적**으로 거동 (서로 운동량에 영향 미치지 않으며, 인력/척력도 작용하지 않음)

입자간의 충돌은 완전 단성이며, 벽과 입차의 충돌은 **압력**이 됨.

[예제1]

Magnesium nitride는 마그네슘 금속과 질소 기체가 반응해서 생긴다

1) 마그네슘 35 g과 질소 15g이 반응하면 magnesium nitride는 몇 g 생기는가?

2) 반응 후 excess reactant는 몇 g 남는가?

[풀이]

1)

마그네슘은 alkaline earth metal이고, 따라서 마그네슘과 실소 반응식은 다음과 같다.

$Mg + N_2 \rightarrow Mg_3N_2$

분자량을 맞추면 다음과 같다.

$3Mg + N_2 \rightarrow Mg_3N_2$

주기율표에서 마그네슘과 질소의 원자량은 다음과 같다.

Mg : 24.305 g/mol, N : 14.007 g/mol ($N_2$ : 28.014 g/mol)

**N2는 원자량을 2배해야 함을 주의!

Mg 35 g은 35 g / 24.305 g/mol = 1.440 mol 이고,

$N_2$ 15 g 은 15 g / 28.013 g/mol = 0.5355 mol이다.

Mg와 $N_2$는 3:1의 몰비로 반응한다.

0.5355*3=1.6065이므로, $N_2$는 excess, Mg는 limiting이다.

$Mg_3N_2$의 분자량은 24.305*3+14.007*2=100.929 g/mol이다.

Mg는 모두 반응하고, Mg와 $Mg_3N_2$는 3:1의 몰비로 반응한다.

따라서 $Mg_3N_2$는 1.440/3 = 0.48 mol이고,

(100.929 g/mol)(0.48 mol)= 48.45 g이다.

2)

$N_2$는 Mg 1.440 mol의 1/3만 반응하고 나머지는 남는다.

0.5355 mol - 1.440/3 mol = 0.0555 mol

(0.0555 mol)(28.014 g/mol)=1.55 g

질산(nitric acid) 2.5 kg이 엎질러졌다. $Na_2 CO_3$ 2 kg이 빠르게 부었고, $CO_2$가 반응 결과로 방출되었다. $Na_2 CO_3$는 모든 산을 중화시키기에 충분히 사용되었는가?

[풀이]

중화시키기에 충분한 양인지 보려면 반응하는 2개 분자의 mol이 같으면 된다.

$HNO_3 + Na_2 CO_3 \rightarrow  NaNO_3 + H_2O + CO_2$

$2HNO_3 + Na_2 CO_3 \rightarrow  2NaNO_3 + H_2O + CO_2$

### Balancing Redox Reaction

Redox는 산화환원(Reduction-Oxidation) 반응이다.

산화(Oxidation) 과정은 전자를 잃고, 환원(Reduction)과정은 전자를 얻는다.

산화과정에서 잃은 전자 수와 환원과정에서 얻은 전자 수는 같아야 한다.

[순서]

1) oxidation number를 확인하고, half-reaction을 작성한다.

oxidation number는 얻을 수 있는 전자 수 있다.

다음을 참고한다.

수소: 비금속과 결합하면 +1, 금속과 결합하면 -1

산소 : 대부분 혼합물에서 -2, 가끔 -1(peroxide, $O_2^{2-}$), 아주가끔 $-1/2 (superoxide, $O_2^{-}$), F와 결합되면 양수

Halogen : F는 항상 -1, 다른 halogen(산소 및 다른 halogen과 결합되지 않은 경우) -1

2) 각 half-reaction을 balancing 한다. (O, H는 제외)

3) 각 half-reaction에 전자를 더하여 charge를 balancing 한다.

4) 두개 harf-reaction의 전자 수를 동일하게 맞춘 후(양 변에 상수 곱) 두개의 reaction을 합친다.

5) $H_2O$를 추가해 O와 H를 balancing 한다. 

6) acidic인 경우 $H^+$를 더해 H를 balancing 한다.

7) basic인 경우  $H^+$를 더해 H를 balancing한 후 동일한 수의 $OH^-$를 양쪽에 더한다.

[예제]

$Br_2$는 용해물 속에서 bromide ion과 bromate ion으로 분해된다. half-reaction method를 사용해 다음 반응을 balancing 하여라.

$Br_2(l)\rightarrow Br^{-}(aq) + BrO_3^{-}$

1) Oxidation number $Br_2$ : 0, $Br^{-}$ : -1, $BrO_3^{-}$ 속의 Br : 산소는 -2이며, 3개이므로 총 -6이고, 전체 분자가 -1이 되기 위해서 Br은 +5가 되어야 한다.

2) 산화 환원 과정을 쓰면 다음과 같다.

$Br_2 + 2e^{-} \rightarrow 2Br^{-} $

$Br_2 \rightarrow 10e^{-} + 2BrO_3^{-}$

3) 두개 반응의 전자수를 같게 해주기 위해 환원 반응 양변에 5를 곱한 후 두개 반응을 더하면,

$6Br_2 + 10e^{-} \rightarrow 10Br^{-} + 10e^{-} + 2BrO_3^{-}$

4) 양변의 전자는 동일하므로 삭제하고, 단순화 하면,

$3Br_2 \rightarrow 5Br^{-} + BrO_3^{-}$

5) $H_2O$를 추가해 O를 balancing 하면,

$3Br_2 + 3H_2O \rightarrow 5Br^{-} + BrO_3^{-}$

6) basic 과정이므로 $H^+$를 추가해 H를 balancing하고 양변에 동일한 양의 $OH^-$를 추가하면,

$3Br_2 + 3H_2O \rightarrow 5Br^{-} + BrO_3^{-} + 6H^{+}$

$3Br_2 + 3H_2O + 6OH^- \rightarrow 5Br^{-} + BrO_3^{-} + 6H^{+}+ 6OH^-$

7) 우변의 $6H^{+}+ 6OH^-$를 $6H_2O$로 변경 후, 공통된 부분을 삭제하면,

$3Br_2 + 6OH^- \rightarrow 5Br^{-} + BrO_3^{-} + 3H_2O$

## equilibrium constants

화학 반응 $aA + bB \rightarrow cC + dD$에서 평형상수(equilibrium constant) $K(K_{eq}, K_c)$는 다음과 같이 정의됨

$K=\frac{a(C)^c a(D)^d}{a(A)^a a(B)^b}$

a는 activity를 나타냄.

보통 activity는 M 내의 concentration으로 대체함

$K=\frac{[C]^c [D]^d}{[A]^a[B]^b}$

K는 얼마나 반응이 잘 진행되는지를 의미함.

K가 크면 반응이 거의 완벽함을 의미하고, 작으면 반응이 잘 일어나지 않음을 의미

대부분의 반응은 가역이지만 매우 큰 K 값 ($>10^10$)인 경우 비가역이라고 부름

K는 무차원 수임

K는 reactant 대비 product의 열역학적 안정성을 나타냄

K는 반응 속도를 의미하지는 않음
