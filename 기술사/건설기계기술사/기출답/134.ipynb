### **1. 도로포장용 건설기계 3가지를 제시하고 특징에 대하여 설명하시오.**  
#### (1) 아스팔트 피니셔  
- **특징**: 자동 높이 조절과 균일한 두께로 포설 가능.  
- **용도**: 도로 표층 포설.  

#### (2) 진동 롤러  
- **특징**: 진동과 하중을 이용한 다짐 성능 우수.  
- **용도**: 포설된 아스팔트 및 기반층 다짐.  

#### (3) 콜드 플래너(밀링기)  
- **특징**: 기존 포장을 제거하며 재활용 가능.  
- **용도**: 유지보수 및 재포장 작업.

#### 실무사례  
- 서울 강남대로 확장공사에서 아스팔트 피니셔와 진동 롤러를 사용하여 도로 평탄성을 ±3mm 이내로 유지함.  
- 인천공항 활주로 재포장 공사에서 밀링기를 사용해 30%의 재료를 재활용함.

---

### **2. 건설기계 중 굴착기의 작업효율 저하 요인 및 대책**  
#### (1) 작업효율 저하 요인  
1. **버킷 마모**로 인한 굴착력 저하.  
2. **유압 계통 문제**로 동력 손실.  
3. **작업 환경**(연약 지반, 장애물 등)으로 작업 속도 저하.  

#### (2) 대책  
1. 버킷 교체 및 정기 점검.  
2. 유압 오일 및 실린더 유지 관리.  
3. 적절한 장비와 보조장치 사용.  

#### 실무사례  
- 경기도 물류센터 터파기 공사에서 유압 실린더 누유 발생 후, 패킹 교체로 작업량이 30% 회복됨.  
- 해안 매립지 작업 시 크롤러 폭을 확장해 장비 침하 방지 및 작업 효율을 20% 향상.

### **3. 공사 시행 시 공정관리 4단계(LevelⅠ~Ⅳ)에 대하여 설명하시오.**  
#### (1) LevelⅠ: 프로젝트 마스터 계획  
- 주요 공정(착공, 골조 완료, 준공) 수립.  

#### (2) LevelⅡ: 세부 공정 계획  
- 각 공정을 세부 작업 단위로 구분.  

#### (3) LevelⅢ: 작업 단위별 관리  
- 작업별 자원 배치 및 효율적 관리.  

#### (4) LevelⅣ: 일일 작업 관리  
- 작업 진행 상황 점검 및 수정.  

#### 실무사례  
- 서부간선도로 지하화 공사에서 LevelⅢ를 활용해 골조공사를 철근 배근, 콘크리트 타설 등으로 세분화하여 공기를 2주 단축함.

### **4. 배관공사 중 부단수 차단(Line Stopping)에 대하여 설명하시오.**  
#### (1) 정의  
- 유체 흐름을 유지하며 특정 구간의 작업을 수행하는 기술.  

#### (2) 특징  
- 핫탭(Hot Tapping) 장비 사용으로 공정 연속성 확보.  
- 밸브 설치 없이 배관 일부를 차단 가능.  

#### 실무사례  
- 수도관 복구공사에서 부단수 차단 기술을 사용해 수돗물 공급 중단 없이 복구 완료.

2. 부드러운 동력 전달.  
3. 과부하 방지.  
4. 유지보수 용이.  
5. 내구성 우수.  

#### 실무사례

### **6. 펌프의 상사법칙 중 유량에 관한 상사법칙을 아래의 인자로 설명하시오.**  
#### (1) 상사법칙  
- 유량 (\(Q\))은 회전수 (\(N\))와 임펠러 직경 (\(D\))에 비례:

  \]$  

#### 실무사례

### **13. 건설기계 재료의 취성파괴를 설명하고 단면적 변화를 도시하시오.**

#### (1) 취성파괴의 정의  
- 재료가 외부의 충격 또는 하중을 받을 때, 변형 없이 갑작스럽게 파괴되는 현상.  
- 일반적으로 낮은 온도, 높은 하중 속도, 결함이 있는 재료에서 발생.  

#### (2) 발생 원인  
1. **재료 내부의 결함**: 기공, 불순물, 미세 균열 등.  
2. **저온 환경**: 온도가 낮아지면 재료의 연성이 감소.  
3. **응력집중**: 모서리나 결합부에서 응력이 집중되면 균열이 발생.  

#### (3) 특징  
- 전단 변형 없이 파괴.  
- 파괴 단면이 매끄럽고 결정립 구조가 드러남.  
- 주로 취성 재료(주철, 고탄소강 등)에서 발생.  

---

#### (4) 단면적 변화  
취성파괴는 재료의 단면적 변화가 거의 없으며, 파괴 단면은 매끄럽고 연성파괴와는 달리 변형 흔적이 나타나지 않습니다.  
아래는 연성파괴와 취성파괴의 단면적 변화를 비교한 도시입니다.

1. **연성파괴**: 단면적이 눈에 띄게 줄어드는 현상이 보임.  
2. **취성파괴**: 단면적이 거의 변하지 않고 원래의 크기와 유사.

---

#### 단면적 변화 도시  

1. **연성파괴**:  
   - 단면이 '컵과 원뿔' 형태로 줄어듦.  
   - 파단 전 연신율이 커짐.  

2. **취성파괴**:  
   - 단면적 변화 없음.  
   - 단면이 매끄럽고 평면적.
