## 원심펌프의 유량 및 양정에 관한 상사법칙(affinity laws)을 이용하여 비속도(specific speed)를 유도

---

### 1. 상사법칙 개요

원심펌프의 형상(기하학적 형상)이 동일할 때, 회전수(n), 임펠러의 대표 치수(D), 유량(Q), 양정(H) 사이에는 다음과 같은 상사관계가 성립합니다.

- **유량 상사법칙**  
  $ Q \propto nD^3 $
    
- **양정 상사법칙**  
  $ H \propto n^2D^2 $
  

여기서 $ Q $는 유량, $ H $는 양정, $ n $은 회전수, $ D $는 임펠러의 대표 치수입니다.

---

### 2. 무차원수 및 계수 도입

펌프 성능을 나타내는 무차원수로 다음과 같은 유량계수와 양정계수를 정의할 수 있습니다.

- **유량 계수 (Capacity Coefficient, $\phi$)**  
 
  $\phi = \frac{Q}{nD^3}$
  
- **양정 계수 (Head Coefficient, $\psi$)**  
  
  $\psi = \frac{H}{n^2D^2}$
  

이 계수들은 서로 상수(설계에 따라 결정되는 상수 $ C_Q $, $ C_H $로 나타낼 수 있으므로,

$Q = C_Q \, nD^3 \quad \text{및} \quad H = C_H \, n^2D^2.$


---

### 3. 비속도 \( N_s \) 유도

비속도는 펌프의 운전점(최고 효율점)에서의 유량과 양정을 고려하여, 임펠러의 지름(또는 형상)과 무관하게 펌프 형식을 결정하기 위한 무차원수입니다. 일반적으로 실용적 비속도는 다음과 같이 정의됩니다.


$N_s = n \frac{\sqrt{Q}}{H^{3/4}}$


이제 상사법칙을 이용하여 위 식이 왜 상수(즉, 동일 상사 조건 하에서 일정한 값)가 되는지 유도해 보겠습니다.

1. 유량 식에서 임펠러 치수를 풀면,
   
   $D = \left(\frac{Q}{C_Q \, n}\right)^{1/3}.$
   
2. 이를 양정 식에 대입하면,
   
   $H = C_H \, n^2 \left(\frac{Q}{C_Q \, n}\right)^{2/3} 
   = C_H \, n^2 \cdot \frac{Q^{2/3}}{C_Q^{2/3} \, n^{2/3}}
   = \frac{C_H}{C_Q^{2/3}} \, Q^{2/3} \, n^{4/3}.$
   
3. 양정을 \(3/4\) 승하면,
   
   $H^{3/4} = \left(\frac{C_H}{C_Q^{2/3}}\right)^{3/4} \, Q^{1/2} \, n.$
   
4. 그러면 비속도 식은,
   $
   N_s = n \frac{Q^{1/2}}{H^{3/4}}
   = n \frac{Q^{1/2}}{\left(\frac{C_H}{C_Q^{2/3}}\right)^{3/4} \, Q^{1/2} \, n}
   = \left(\frac{C_Q^{2/3}}{C_H}\right)^{3/4},
   $
   즉, 설계상 일정한 상수로 결정됩니다.

이로써 동일 상사 조건(즉, 기하학적, 동유동적 상사가 만족될 때)에서는 \( N_s \)가 임펠러 치수나 회전수에 의존하지 않고 일정한 값이 됨을 알 수 있습니다.

---

### 4. 결론

따라서 원심펌프의 비속도는 상사법칙에 따라  

$\boxed{N_s = n\,\frac{\sqrt{Q}}{H^{3/4}}}$

로 유도되며, 이는 펌프의 형식 및 임펠러 설계 결정 시 중요한 기준으로 활용됩니다.
