## 발전플랜트에서 주로 사용되는 가스터빈과 증기터빈의 원리 및 열역학적 사이클

---

## 1. 가스터빈 (Gas Turbine)

### 원리
가스터빈은 압축기, 연소실, 터빈의 세 주요 구성요소로 이루어집니다.  
- **압축기(Compressor):** 대기 중의 공기를 고압으로 압축합니다.  
- **연소실(Combustor):** 압축된 공기와 연료가 혼합되어 연소가 일어나며, 이 과정에서 고온·고압의 가스가 생성됩니다.  
- **터빈(Turbine):** 연소실에서 발생한 고온·고압의 가스가 터빈을 팽창하면서 동력으로 변환됩니다. 이 동력은 압축기를 구동하거나 발전기를 회전시켜 전력을 생산하는 데 사용됩니다.

### 열역학적 사이클 – 브레이튼 사이클 (Brayton Cycle)
가스터빈의 이상적인 사이클은 브레이튼 사이클로, 다음과 같은 4가지 주요 과정을 포함합니다.

1. **단계 1 → 2: 등엔트로피 압축 (Adiabatic Compression)**  
   - 공기가 압축기를 통과하면서 압력과 온도가 상승하며, 이 과정은 이상적으로 가역적 단열과정(등엔트로피 과정)으로 진행됩니다.

2. **단계 2 → 3: 등압 가열 (Isobaric Heat Addition)**  
   - 압축된 공기가 연소실에서 연료와 혼합되어 연소하면서 온도가 급격히 상승합니다. 이때 압력은 상대적으로 일정하게 유지됩니다.

3. **단계 3 → 4: 단열 팽창 (Adiabatic Expansion)**  
   - 고온·고압의 연소 가스가 터빈 내에서 팽창하며 일을 수행합니다. 이상적으로 이 과정도 등엔트로피 과정입니다.

4. **단계 4 → 1: 등압 방열 (Isobaric Heat Rejection)**  
   - 터빈을 통과한 후 배기가스는 대기와 열교환을 통해 열을 방출하면서 원래 상태로 복귀합니다.

실제 운전에서는 압축기와 터빈에서의 내부 손실, 압력강하, 그리고 재생(회수) 사이클이나 인터쿨링, 재열 등의 보완 사이클이 적용되어 효율을 높이게 됩니다.

## 2. 증기터빈 (Steam Turbine)

### 원리
증기터빈은 보일러나 HRSG(Heat Recovery Steam Generator)에서 발생한 고온·고압의 증기를 이용하여 터빈을 회전시킵니다.  
- **보일러/HRSG:** 연료를 연소하거나 가스터빈의 배기가스로 물을 가열하여 증기를 발생시킵니다.  
- **증기터빈:** 발생된 증기가 팽창하면서 터빈 블레이드를 회전시켜 기계적 에너지를 생산합니다.  
- **응축기 및 펌프:** 터빈을 통과한 증기를 응축하여 다시 물로 만들고, 이 물을 보일러로 재순환시키는 Rankine 사이클의 일부로 사용됩니다.

### 열역학적 사이클 – 랭킨 사이클 (Rankine Cycle)
증기터빈의 이상적인 사이클은 랭킨 사이클로, 주요 구성 단계는 다음과 같습니다.

1. **펌프(Feedwater Pump):** 응축기에서 나온 물을 고압으로 압축하여 보일러로 공급합니다.
2. **보일러/HRSG:** 압축된 물이 고온·고압의 증기로 변환됩니다.
3. **증기팽창 (Turbine Expansion):** 고압 증기가 터빈에서 팽창하며 일을 수행하여 전력으로 변환됩니다.
4. **응축기(Condenser):** 터빈을 통과한 저압 증기를 냉각시켜 다시 물로 응축시킵니다.

이 사이클에서는 응축기에서의 응축과 펌핑 과정이 사이클 효율에 중요한 영향을 미치며, 복수식 터빈(Condensing Turbine)과 배압식 터빈(Back Pressure Turbine) 등 운전 방식에 따라 세부 특성이 달라집니다.  

---

## 3. 복합 사이클 발전 시스템 (Combined Cycle Power Plant)
복합 사이클 발전소에서는 가스터빈과 증기터빈을 결합하여 운영합니다.  
- **가스터빈:** 브레이튼 사이클을 따르며 전력을 생산한 후, 고온의 배기가스를 HRSG로 보내어 증기를 발생시킵니다.
- **증기터빈:** 발생된 증기를 이용해 추가 전력을 생산합니다.

이와 같이 두 사이클을 결합하면 단일 사이클에 비해 열효율이 크게 향상되어 55~60% 정도의 효율을 달성할 수 있습니다.  

---

## 결론
가스터빈과 증기터빈은 각각 브레이튼 사이클과 랭킨 사이클을 기반으로 동작하며, 이들 사이클은 발전플랜트의 효율과 성능에 직결됩니다.  
- 가스터빈은 빠른 반응속도와 높은 출력밀도, 그리고 재생 사이클 등으로 효율을 높일 수 있는 장점이 있으며,  
- 증기터빈은 응축-펌프 사이클을 통한 물의 재순환으로 안정적인 운전을 지원합니다.

두 터빈의 특성을 이해하고, 각각의 열역학 사이클에 따른 손실 및 보완기술을 고려하는 것이 발전플랜트 설계 및 운전에 있어서 매우 중요합니다.
