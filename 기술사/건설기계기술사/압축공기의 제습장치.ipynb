문제 : 압축 공기의 제습장치에 대하여 설명하시오.

I. 압축공기 제습장치의 정의

압축공기 중의 포화수증기를 물리적 또는 인위적인 방법으로 제거하여 주는 장치

[압축공기를 제습하는 이유]

주된 이유 : 압축공기에서 발생가능한 응축수 제거

상세 이유

(1) 공기의 온도하락으로 발생되는 응축수로 인한 관련기기, 계측기, 작동기기, 정밀측정기 등의 부식과 오작동을 방지하고 수명을 연장

(2) 겨울철 동파방지

(3) 초건조공기를 필요로 하는 첨단산업(반도체, 원료, 의료, 식품) 

(4) 압축공기 중에 함유되어 있는 수분을 제거하여 압축공기의 질을 높여 에어라인을 보호

(5) 압축공기가 소요되는 부품 및 말단기기의 TROUBLE을 방지하고 수명을 보호하기 위함

[응축수 발생 순서]

(1) 압축기는 대기 공기를 주로 사용하므로, 압축기에서 생산 공급되는 압축공기에는 다량의 수증기가 포함됨

(2) 이 수증기의 양은 대기의 온도, 습도, 압축된 압력에 비례하여 배수로 높은 온도의 압축 공기 중에 존재함

(3) Line을 흐르는 도중 온도가 떨어지면(외기의 영향 또는 유속 등의 원인) 응축수로 변화함

II. 압축공기 제습장치 종류

(1)냉동식 에어드라이어

냉동에 의한 제습은 공기조화에서 가장 일반적으로 쓰이고 있으며, 냉방등의 경우는 편리 하지만, 온도가 내려가면 지장이 있을 때는 재열을 요하므로 비경제적이다. 또한 소요 노 점온도가 낮을 때는 냉각면에 응축수가 빙결되어 공기의 흐름이 차단되므로 그에 대한 대 비가 요망된다. 

냉동식 제습법의 원리는 공기를 노점온도 이하로 냉각하기 위해 냉동기를 사용하기 때문에 냉동식 제습장치라고도 한다. 

냉동기에 압축된 냉매를 팽창시켜 열교환기에 압축공기를 극도로 4~10℃로 냉각시켜서 공 기중의 수분을 응축시켜 제습하는 방식으로써 압력하 노점(DEW POINT) 4℃까지 제습할 수 있다.

냉동식 드라이어의 경우 주위온도와 입구온도에 의한 제습율이 상당히 많이 생길 수 있으 며, 현장 AIR 배관의 방법 및 길이와 현장내 주위 온도에 의해 또다시 배관내부에 수분이 발생할 수 있다.

냉각 제습으로 발생하는 응축수는 흡착식 제습의 제습량에 비해 10배 가까이 되므로 Drain 분리가 극히 중요하다.

Mist상의 드레인은 완전히 분리하는 것이 여간 어렵지 않기 때문에 분리효율이 높은 Drain Separator를 선정하는 것이 중요하며 드레인을 외부로 배출하는 Air Trap도 극히 중요하다.

장점

- 연속운전이 가능하다.

- 흡착제 등 소모품이 불필요하다.

- 재생공기 및 전열이 불필요하다.

- 작은 마력일 경우 구입비와 운전경비가 낮다.

단점

- 낮은 노점을 얻을 수 없다. 

- 냉매가 누설될 경우 제습 효율이 저하된다.

- 동절기에 말단 배관라인에 수분이 발생하기 때문에 보온이 필요하다.

- 소형은 가격이 싸지만 대형은 용해식, 흡착식보다 유지비, 구입비등 모두 비싸다.

- 4℃ 이하의 노점을 얻을 수 없으며 동절기에 영하의 기온에서는 효과가 미비하다.

- 운전비용은 Heatless보다 싸지만 장치의 복잡성과 운전비용과 

보수방법의 특수성으로 유지비가 많이든다.

(2) 용해식 에어드라이어

용해식 제습법은 흡습성이 있는 고체 또는 액체에 공기를 접촉시켜 제습하는 것이며, 적절히 설계하면 필요로 하는 온․습도가 동시에 얻어진다. 일반적으로 냉동식으로는 곤란한 저 노점이 필요할 때 쓰인다. 

공기를 Tank내의 용해제로 통과시키면 용해제가 수분을 흡수하며 녹아내리는 방식으로 에어의 사용시간과 수분함유량에 따라 용해제의 충진 주기가 달라진다.

장점 : 초기 투자비가 흡착식보다 저렴하다.

무동력으로 운전유지비가 전혀 들지 않는다.

탱크와 용해제로만 구성되어 고장이 없어 수명이 반영구적이다.

냉동식보다 낮은 저 노점을 얻을 수 있어 수분제거율이 높다.

단점 : 노점이 입구 공기온도에 따라 변한다.

수분을 흡수하여 녹아내린 용해제는 배출시켜야 하며,

용해제는 년2~3회 보충시켜주어야 하는 단점이 있다.

(3) 흡착식 에어드라이어

고체흡착제(알루미나 겔)을 2개의 타워에 충전하여 서로 반복 연속하여 제습과 재생을 하는 방식으로서 압력하노점 -40℃ 이하까지 제습할 수 있는 에어 드라이어이다.

흡착식 드라이어의 경우 드라이어 전단에 유분을 0.01ppm 까지 제거할 수 있는 Air Filter를 설치하여야 한다.

흡착은 흡착제의 표면, 즉 상의계면에 기체의 분자가 정체하는 것으로서 일반적으로 수분이 흡착제의 모관 내벽에 부착하는 표면흡착과 내부에 응축하는 모관 응축에 의한 것이라고 말하고 있으며, 표면흡착은 화학흡착이라고도 말할 수 있다. 

흡착식 제습장치는 수분 흡착제가 담겨진 두 개의 통과 이를 교대로 사용하기 위한 절환변, 그리고 흡착된 수분을 탈착시키기 위한 재생 시스템으로 구별된다.

흡착법은 고형흡착제를 사용하므로, 열효율이 액체에 의한 흡습법 보다 낮으나, 매우 낮은 노점까지 제습할 수 있어 초건조 공기를 필요로 하는 공정에 쓰인다.

히터레스 흡착식 드라이어

: 구입 가격은 가장 저렴하며 흡착제 재생시 15 ~ 17%까지 압축공기가 소모되는 단점이 있다.

히터외장형 흡착식 드라이어

: 구입 가격이 고가이며 흡착제 재생시 5 ~ 7% 정도가 소모되며, 재생시 히터 가열로인한 전력이 소비된다. 이 경우 흡착제를 충분히 가열 재생함으로써 최저노점 요구에 가장 적합하며 가장 많이 사용하는 시스템이다.

흡착식 제습장치는 재생하는 방법에 따라 다음과 같이 여러 종류로 구분되고 있다.
