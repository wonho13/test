```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mendelbrot(x0,x1,y0,y1):
    n_grid=1000;
    x,y=np.meshgrid(np.linspace(x0,x1,n_grid),\         
                    np.linspace(y0,y1,n_grid));
    c=x+y*1j
    
```

```python
c=x+y*1j
```

```python
n_iter=100;
M=np.zeros([n_grid,n_grid])
z=c
for i in range(n_iter):
    z=z**2+c
    M[np.abs(z)>2]=i
```
